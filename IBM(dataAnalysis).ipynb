# **Binning**

https://www.linkedin.com/posts/omnia-mohamed-ghazy-380388299_google-colab-activity-7356414523861344257-_BhD?utm_source=social_share_send&utm_medium=member_desktop_web&rcm=ACoAAEgi9BABW1lf-LchtTby7Y-AFZSDgZ_wTIg

```python
age_bins = [18, 25, 35, 45, 55, 65]
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-65']

df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=True, include_lowest=True)

print("Age بعد Binning:")
print(df[['Age', 'Age_Group']].head())
```

```python

df['Income_Quartile'] = pd.qcut(df['MonthlyIncome'], q=4, labels=['Q1_Low', 'Q2_Medium', 'Q3_High', 'Q4_VeryHigh'], duplicates='drop')

print("\nMonthlyIncome بعد Binning:")
print(df[['MonthlyIncome', 'Income_Quartile']].head())
```

```python
years_bins = [0, 1, 5, 10, 20, df['YearsAtCompany'].max() + 1]
years_labels = ['<1 Year', '1-5 Years', '6-10 Years', '11-20 Years', '20+ Years']

df['YearsAtCompany_Group'] = pd.cut(df['YearsAtCompany'], bins=years_bins, labels=years_labels, right=True, include_lowest=True)

print("\nYearsAtCompany بعد Binning:")
print(df[['YearsAtCompany', 'YearsAtCompany_Group']].head())
```
